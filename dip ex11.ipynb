```python
# Import the necessary packages
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Create the Text using cv2.putText
text_img = np.zeros((100,500),dtype = 'uint8')
font = cv2.FONT_HERSHEY_SIMPLEX 
img=cv2.putText(text_img," Sangeetha",(5,70),font,2,(255),5,cv2.LINE_AA)
plt.title("Original Image")
plt.imshow(img,'bone')
plt.axis('off')
```

```python
# Create the structuring element
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(11,11))
```

```python
# Use Opening operation
open_image = cv2.morphologyEx(img,cv2.MORPH_OPEN,kernel)
plt.title("Opening")
plt.imshow(open_image, 'twilight_r')
plt.axis('off')
```

```python
# Use Closing Operation
close_image = cv2.morphologyEx(img,cv2.MORPH_CLOSE,kernel)
plt.title("Closing")
plt.imshow(close_image,'bone')
plt.axis('off')
```
